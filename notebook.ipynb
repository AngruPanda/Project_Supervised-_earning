## Введение

Интернет-магазин «В один клик» продаёт разные товары: для детей, для дома, мелкую бытовую технику, косметику и даже продукты. Отчёт магазина за прошлый период показал, что активность покупателей начала снижаться. Привлекать новых клиентов уже не так эффективно: о магазине и так знает большая часть целевой аудитории. Возможный выход — удерживать активность постоянных клиентов. Сделать это можно с помощью персонализированных предложений.

«В один клик» — современная компания, поэтому её руководство не хочет принимать решения просто так — только на основе анализа данных и бизнес-моделирования. У компании есть небольшой отдел цифровых технологий, и вам предстоит побыть в роли стажёра в этом отделе. 

### Задача

Разработать решение, которое позволит персонализировать предложения постоянным клиентам, чтобы увеличить их покупательскую активность.

### План работ

Продумывать подход к решению этой задачи вам не придётся — руководитель отдела его уже сформировал:
1. Нужно промаркировать уровень финансовой активности постоянных покупателей. В компании принято выделять два уровня активности: «снизилась», если клиент стал покупать меньше товаров, и «прежний уровень».
2. Нужно собрать данные по клиентам по следующим группам:
    - Признаки, которые описывают коммуникацию сотрудников компании с клиентом.
    - Признаки, которые описывают продуктовое поведение покупателя. Например, какие товары покупает и как часто.
    - Признаки, которые описывают покупательское поведение клиента. Например, сколько тратил в магазине.
    - Признаки, которые описывают поведение покупателя на сайте. Например, как много страниц просматривает и сколько времени проводит на сайте.
    
Представим группы признаков (вместе с целевым) в виде диаграммы — такую визуализацию ещё называют диаграммой Исикавы.

- Нужно построить модель, которая предскажет вероятность снижения покупательской активности клиента в следующие три месяца.
- В исследование нужно включить дополнительные данные финансового департамента о прибыльности клиента: какой доход каждый покупатель приносил компании за последние три месяца.
- Используя данные модели и данные о прибыльности клиентов, нужно выделить сегменты покупателей и разработать для них персонализированные предложения.

Руководство одобрило описание решения, и вам, как специалисту по DS, нужно его реализовать.

### Описание данных

Данные для работы находятся в нескольких таблицах. Нажмите на название файла, чтобы его скачать.

<font color='red'>market_file.csv</font>

Таблица, которая содержит данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении.
* <font color='green'>id</font> — номер покупателя в корпоративной базе данных.
* <font color='green'>Покупательская активность</font> — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень».
* <font color='green'>Тип сервиса</font> — уровень сервиса, например «премиум» и «стандарт».
* <font color='green'>Разрешить сообщать</font> — информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель.
* <font color='green'>Маркет_актив_6_мес</font> — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента.
* <font color='green'>Маркет_актив_тек_мес</font> — количество маркетинговых коммуникаций в текущем месяце.
* <font color='green'>Длительность</font> — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте.
* <font color='green'>Акционные_покупки</font> — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев.
* <font color='green'>Популярная_категория</font> — самая популярная категория товаров у покупателя за последние 6 месяцев.
* <font color='green'>Средний_просмотр_категорий_за_визит</font> — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца.
* <font color='green'>Неоплаченные_продукты_штук_квартал</font> — общее число неоплаченных товаров в корзине за последние 3 месяца.
* <font color='green'>Ошибка_сервиса</font> — число сбоев, которые коснулись покупателя во время посещения сайта.
* <font color='green'>Страниц_за_визит</font> — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца.

<font color='red'>market_money.csv</font>

Таблица с данными о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом.

* <font color='green'>id</font> — номер покупателя в корпоративной базе данных.
* <font color='green'>Период</font> — название периода, во время которого зафиксирована выручка. Например, 'текущий_месяц' или 'предыдущий_месяц'.
* <font color='green'>Выручка</font> — сумма выручки за период.

<font color='red'>market_time.csv</font>

Таблица с данными о времени (в минутах), которое покупатель провёл на сайте в течение периода.
* <font color='green'>id</font> — номер покупателя в корпоративной базе данных.
* <font color='green'>Период</font> — название периода, во время которого зафиксировано общее время.
* <font color='green'>минут</font> — значение времени, проведённого на сайте, в минутах.

<font color='red'>money.csv</font>

Таблица с данными о среднемесячной прибыли покупателя за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю.
* <font color='green'>id</font> — номер покупателя в корпоративной базе данных.
* <font color='green'>Прибыль</font> — значение прибыли.

## Объединение таблиц

Для построения модели машинного обучения нужно объеденить таблицы в одну. Но в двух таблицах информация расположена не совсем удобно, поэтому необходимо для каждого месяца сделать столбцы с выручкой и временем, для этого сделаем сводные таблицы.

```python
market_money_temp = pd.pivot_table(market_money, columns = 'Период', index='id', values='Выручка')
market_money_temp.columns = ['предыдущий_месяц_выручка', 'препредыдущий_месяц_выручка', 'текущий_месяц_выручка']
market_money_temp = market_money_temp.reset_index()

market_money_temp
```

```python
market_time_temp = pd.pivot_table(market_time, columns=['Период'], index='id', values='минут')
market_time_temp.columns = ['предыдущий_месяц_время', 'текущий_месяц_время']
market_time_temp = market_time_temp.reset_index()

market_time_temp
```

Теперь объеденим все три таблицы по столбцу с id

```python
full_market = market_file.merge(market_time_temp.merge(market_money_temp, on='id'), on='id').dropna()

full_market
```

## Подбор модели

Подбор модели и гиперпараметров будем производить с помощью пайплайнов. Категориальные признаки будет кодировать с помощью OneHotEncoder и OrdinalEncoder. Для кодирования количественных признаков будем использовать один из этих кодировщиков StandardScaler, MinMaxScaler, RobustScaler. Перебираемые модели с их гиперпараметрами:
- DecisionTreeClassifier
    - max_depth: 2-5
    - max_features: 2-5
- KNeighborsClassifier
    - n_neighbors: 2-5
- LogisticRegression
    - C: 1-5
- RandomForest
    - n_estimators: range(80, 100, 4)
    - max_depth: 2-5
    - max_features: 2-5
- SVC
    - C: 1-5
    
Для самого перебора будем использовать RandomizedSearchCV. В качестве метрики отбора лучшей модели возьмем ROC-AUC, так как она оценивает качество решений модели в наиболее общем виде, учитывая все возможные пороговые значения, также такая метрика хорошо работает в случае, когда объектов с одним значением целевого признака больше, чем с другими, а у нас как раз таки такая ситуация в данных

```python
X_train, X_test, y_train, y_test = train_test_split(
    full_market.drop(['Покупательская_активность', 'id'], axis=1),
    full_market['Покупательская_активность'],
    test_size = TEST_SIZE, 
    random_state = RANDOM_STATE)

ohe_columns = [ 'Разрешить_сообщать']
ord_columns = ['Маркет_актив_тек_мес', 'Популярная_категория', 'Тип_сервиса']
num_columns = ['Маркет_актив_6_мес', 'Маркет_актив_тек_мес', 'Длительность',
               'Акционные_покупки', 'Средний_просмотр_категорий_за_визит', 'Неоплаченные_продукты_штук_квартал',
              'Ошибка_сервиса', 'Страниц_за_визит', 'предыдущий_месяц_время',
               'текущий_месяц_время', 'предыдущий_месяц_выручка', 'препредыдущий_месяц_выручка',
              'текущий_месяц_выручка']
```

```python
ohe_pipe = Pipeline(
    [
            (
                'ohe', 
                OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
            )
    ]
)

ord_pipe = Pipeline(
    [
            (
                'ord',
                OrdinalEncoder(
                    categories=[
                        [3, 4, 5],
                        ['Товары для детей', 'Домашний текстиль', 'Косметика и аксесуары',
                        'Кухонная посуда', 'Техника для красоты и здоровья',
                        'Мелкая бытовая техника и электроника'],
                        ['стандарт', 'премиум']
                    ]
                )
            )
    ]
)
```

```python
data_prep = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', StandardScaler(), num_columns)
    ],
    remainder='passthrough',
    force_int_remainder_cols=False
)
```

```python
pipe_final= Pipeline(
    [
        ('preprocessor', data_prep),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)
```

```python
param_grid = [
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__min_samples_split': range(2, 5),
        'models__max_depth': range(2, 5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), MaxAbsScaler()]  
    },
    
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), MaxAbsScaler()]   
    },

    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE, 
            solver='liblinear', 
            penalty='l1'
        )],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), MaxAbsScaler()]  
    },
    
    {
        'models': [RandomForestClassifier(random_state=RANDOM_STATE)],
        'models__n_estimators': range(80, 120),
        'models__min_samples_split': range(2, 5),
        'models__max_depth': range(2, 5),
        'models__max_features': range(2, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), MaxAbsScaler()]  
    },
    
    {
        'models': [SVC(random_state=RANDOM_STATE, kernel='linear',probability=True)],
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), RobustScaler(), MaxAbsScaler()]  
    }    
]

grid = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5,
    scoring = 'roc_auc', 
    n_jobs=-1,
    random_state=RANDOM_STATE
)
```

```python
grid.fit(X_train, y_train)
```

В результате перебора моделей и гиперпараметров, наилучший результат показала RandomForestClassifier(max_depth=4, max_features=2, n_estimators=99,random_state=42). Посмотрим какие результаты показа лучшие модели с помощью метрики auc_roc.

```python
pd.DataFrame(grid.cv_results_)[['rank_test_score', 'param_models', 'mean_test_score','params']].sort_values('rank_test_score')
```

```python
features =  grid.best_estimator_['preprocessor'].fit_transform(X_test)

probabilities = grid.best_estimator_['models'].predict_proba(features)
probabilities_one = probabilities[:, 1]
print('Площадь ROC-кривой:', roc_auc_score(y_test, probabilities_one))
```

## Интерпритация модели

Теперь проанализируем с помощью метода SHAP какое влияние оказывает каждый признак на результат прогназировния модели

```python
X_train_2 = grid.best_estimator_['preprocessor'].fit_transform(X_train)
X_train_2 = pd.DataFrame(X_train_2, columns = grid.best_estimator_['preprocessor'].get_feature_names_out())

background_sample = shap.kmeans(X_train_2.values, 100)

explainer = shap.KernelExplainer(grid.best_estimator_['models'].predict_proba, background_sample)

X_test_2 = grid.best_estimator_['preprocessor'].transform(X_test)
X_test_2 = pd.DataFrame(X_test_2, columns = grid.best_estimator_['preprocessor'].get_feature_names_out())

shap_values = explainer(X_test_2)
```

```python
shap.plots.beeswarm(shap_values[:, :, 0]) 
```

По диаграмме можно сказать, что наибольший эффект оказывают время на сайте, количество просмотренных страниц и категорий, а также акционные покупки. То есть можно сказать чем больше времени на сайте, просмотрено стрниц и категорий, тем больше вероятность того что интерес клиента сохранится, а вот чем больше акционных покупок, тем меньше вероятность сохранения интереса. 

Малое влияние оказывает остальные признаки такие как популярные категории, ошибки в сервисе, тип подписки и тд.

Из графика можно сделать вывод, что снизилась покупательная активность у пользователей, у которых много покупок по акции.
